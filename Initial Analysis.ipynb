```python
import pandas as pd
import csv
import os
import math
import json
```

```python
def dist(x1,y1,x2,y2):
    return math.sqrt(math.pow((x2-x1),2)+math.pow((y2-y1),2))
```

```python
#Return a list of new information:
# Number of Steps
# Total distance traveled
# Average distance per step
# Time (in minutes) total
# Number of buttons pressed
def get_info(file_name):
    tot = 0
    fl = pd.read_csv("results/"+file_name)
    xs = fl['x']
    ys = fl['y']
    steps = len(xs)
    for i in range(steps-1):
        tot+=dist(xs[i],ys[i],xs[i+1],ys[i+1])
    avg = tot/steps
    time = (fl['time'][len(fl)-1]-fl['time'][0])/1000.0/60
    buttons = fl['sofar'][len(fl)-1]
    return [steps] + [tot] + [avg] + [time] + [buttons]
```

```python
# clear_completition    if true, cC
# clear_progress        if true, cP
# clear_rules           if true, cR
def get_condition(file_name):
    cond = file_name.split("_")[0]
    return [int("cC" in cond), int("cP" in cond), int("cR" in cond)]
    #ret_val = [0,0,0]
    #if("cC" in cond):
    #    ret_val[0] = 1
    #elif ("cP" in cond):
    #    ret_val[1] = 1
    #elif ("cR" in cond):
    #    ret_val[2] = 1
    #return ret_val
```

```python
condition_time = pd.DataFrame(columns=['Clear Completion','Clear Progress','Clear Rules','Steps to Completion','Distance Traveled', 'Average Distance Per Step', 'Time Spent', 'Buttons Pushed'])
for file in os.listdir("results/"):
    if(file.endswith(".csv")):
        #steps = len(pd.read_csv("results/"+file))
        #dist = get_dist("results/"+file)
        condition_time.loc[len(condition_time)] = get_condition(file) + get_info(file)
```
