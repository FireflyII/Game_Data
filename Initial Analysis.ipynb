```python
import pandas as pd
import csv
import os
```

```python
# clear_completition    if true, cC
# clear_progress        if true, cP
# clear_rules           if true, cR
def get_condition(file_name):
    cond = file_name.split("_")[0]
    ret_val = [0,0,0]
    if("cC" in cond):
        ret_val[0] = 1
    elif ("cP" in cond):
        ret_val[1] = 1
    elif ("cR" in cond):
        ret_val[2] = 1
    return ret_val
```

```python
condition_time = pd.DataFrame(columns=['Clear Completion','Clear Progress','Clear Rules','Steps to Completion'])
for file in os.listdir("results/"):
    if(file.endswith(".csv")):
        steps = len(pd.read_csv("results/"+file))
        condition_time.loc[len(condition_time)] = get_condition(file) + [steps]
```
